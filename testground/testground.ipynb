```python
import matplotlib.pyplot as plt

# Daten für die Graphen
x = range(1, 10)
y1 = [xi * 2 for xi in x]
y2 = [xi ** 2 for xi in x]
y3 = [xi ** 0.5 for xi in x]

# Zeichnen der Graphen
plt.plot(x, y1, label='Linear')
plt.plot(x, y2, label='Quadratisch')
plt.plot(x, y3, label='Wurzel')

# Hinzufügen von Titel und Beschriftungen
plt.title('Drei Graphen in einem Diagramm')
plt.xlabel('X-Achse')
plt.ylabel('Y-Achse')

# Anzeigen der Legende
plt.legend()

# Anzeigen des Diagramms
plt.show()
```

```python
import math
from functools import reduce

def lcm(a, b):
    return a * b // math.gcd(a, b)

def lcm_of_list(numbers):
    return reduce(lcm, numbers)

numbers = [19199, 11309, 17621, 20777, 16043,15517  ]
print(lcm_of_list(numbers))  # outputs: 720
```
